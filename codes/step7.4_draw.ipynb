```python
import os
import sys
import pandas as pd
import numpy as np
import scipy.io as scio
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
from matplotlib.pyplot import MultipleLocator
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['pdf.fonttype'] = 42
# matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['ps.usedistiller'] = 'xpdf'
from matplotlib import font_manager
plt.rcParams["font.sans-serif"] = "Arial" 

```

## Draw Dice and Tract Extension Ratio

```python
categories = np.arange(2)
labels = ['left', 'right']
bar_width = 0.35  # 每组柱的宽度

Dice_human_macaque = [0.19581, 0.22732]
Dice_human_marmoset = [0.25486, 0.27122]

group1_positions = categories - bar_width/2 - 0.01
group2_positions = categories + bar_width/2 + 0.01

fig, ax = plt.subplots(figsize=(3,2), dpi=200)
ax.bar(group1_positions, Dice_human_macaque, bar_width, label='macaque', color='#97C6DF')
ax.bar(group2_positions, Dice_human_marmoset, bar_width, label='marmoset', color='#B8B8D9')

ax.set_xlabel('')
ax.set_ylabel('Dice')
ax.set_xticks(categories)
ax.set_xticklabels([f'{labels[i]}' for i in np.arange(len(labels))])
ax.set_ylim(0.1, 0.35)
# ax.legend()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.xticks(fontsize=12)
plt.yticks(fontsize=8)
plt.ylabel('Dice', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
categories = np.arange(2)
labels = ['left', 'right']
bar_width = 0.35  # 每组柱的宽度

ratio_human_macaque = [3.4574, 2.6265]
ratio_human_marmoset = [2.2508, 2.0887]

group1_positions = categories - bar_width/2 - 0.01
group2_positions = categories + bar_width/2 + 0.01

fig, ax = plt.subplots(figsize=(3,2), dpi=200)
ax.bar(group1_positions, ratio_human_macaque, bar_width, label='macaque', color='#97C6DF')
ax.bar(group2_positions, ratio_human_marmoset, bar_width, label='marmoset', color='#B8B8D9')

ax.axhline(y=1, color='#C4C4C4', linestyle='-', linewidth=1.5)

ax.set_xlabel('')
ax.set_ylabel('Dice')
ax.set_xticks(categories)
ax.set_xticklabels([f'{labels[i]}' for i in np.arange(len(labels))])
ax.set_ylim(0.5, 4)
# ax.legend()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.xticks(fontsize=12)
plt.yticks(fontsize=8)
plt.ylabel('Tract extension ratio', fontsize=12)
plt.tight_layout()
plt.show()
```
