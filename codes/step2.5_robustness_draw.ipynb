```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#  txt 
data1 = pd.read_csv('/n02dat01/users/yfwang/MarmosetWM/result/robustness/within_MBM.txt', header=None)
data2 = pd.read_csv('/n02dat01/users/yfwang/MarmosetWM/result/robustness/within_brain_minds.txt', header=None)
data3 = pd.read_csv('/n02dat01/users/yfwang/MarmosetWM/result/robustness/across_cohort.txt', header=None)

#  DataFrame
df = pd.DataFrame({
    'Category': ['MBM to MBM'] * len(data1) + ['Brain/MINDS to Brain/MINDS'] * len(data2) + ['MBM to Brain/MINDS'] * len(data3),
    'Value': pd.concat([data1, data2, data3], ignore_index=True)[0]
})

# 
plt.figure(figsize=(6, 8), dpi=150)
sns.violinplot(x='Category', y='Value', data=df)
plt.title('')
plt.show()
```
