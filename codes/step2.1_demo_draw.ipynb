```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/n02dat01/users/yfwang/MarmosetWM/support_data/marmoset_dataset_minds.csv')

# Pivot table to get counts for each age and gender
age_counts = df.pivot_table(index='Age', columns='Gender', aggfunc='size', fill_value=0)
weight_counts = df.pivot_table(index='Weight', columns='Gender', aggfunc='size', fill_value=0)

# Define the colors
colors = {'F': '#97C6DF', 'M': '#FDA965'}

# Create the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5), dpi=300)

# Plot for Age
age_indices = np.arange(len(age_counts))
axes[0].bar(age_indices, age_counts['F'], label='F', color=colors['F'])
axes[0].bar(age_indices, age_counts['M'], bottom=age_counts['F'], label='M', color=colors['M'])

axes[0].set_xlabel('Age (y)')
axes[0].set_ylabel('Number of marmosets')
axes[0].set_xticks(age_indices)
axes[0].set_xticklabels(age_counts.index)
axes[0].set_ylim([0,35])
axes[0].legend()

# Plot for Weight
weight_indices = np.arange(len(weight_counts))
axes[1].bar(weight_indices, weight_counts['F'], label='F', color=colors['F'])
axes[1].bar(weight_indices, weight_counts['M'], bottom=weight_counts['F'], label='M', color=colors['M'])

axes[1].set_xlabel('Weight (g)')
axes[1].set_ylabel('Number of marmosets')
axes[1].set_xticks(weight_indices)
axes[1].set_xticklabels(weight_counts.index)
axes[1].set_ylim([0,25])
axes[1].legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/n02dat01/users/yfwang/MarmosetWM/support_data/marmoset_dataset_MBM.csv')

# Pivot table to get counts for each age and gender
age_counts = df.pivot_table(index='Age', columns='Gender', aggfunc='size', fill_value=0)
weight_counts = df.pivot_table(index='Weight', columns='Gender', aggfunc='size', fill_value=0)

new_age_row = {'F':0, 'M':0}
age_counts.loc[8] = new_age_row
age_counts.loc[10] = new_age_row
age_counts = age_counts.sort_index()

new_weight_row = {'F':0, 'M':0}
weight_counts.loc[575] = new_weight_row
weight_counts.loc[600] = new_weight_row
weight_counts = weight_counts.sort_index()

# Define the colors
colors = {'F': '#97C6DF', 'M': '#FDA965'}

# Create the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5), dpi=300)

# Plot for Age
age_indices = np.arange(len(age_counts))
axes[0].bar(age_indices, age_counts['F'], label='F', color=colors['F'])
axes[0].bar(age_indices, age_counts['M'], bottom=age_counts['F'], label='M', color=colors['M'])

axes[0].set_xlabel('Age (y)')
axes[0].set_ylabel('Number of marmosets')
axes[0].set_xticks(age_indices)
axes[0].set_xticklabels(age_counts.index)
axes[0].set_ylim([0,10])
axes[0].legend()

# Plot for Weight
weight_indices = np.arange(len(weight_counts))
axes[1].bar(weight_indices, weight_counts['F'], label='F', color=colors['F'])
axes[1].bar(weight_indices, weight_counts['M'], bottom=weight_counts['F'], label='M', color=colors['M'])

axes[1].set_xlabel('Weight (g)')
axes[1].set_ylabel('Number of marmosets')
axes[1].set_xticks(weight_indices)
axes[1].set_xticklabels(weight_counts.index)
axes[1].set_ylim([0,10])
axes[1].legend()

# Show the plots
plt.tight_layout()
plt.show()
```
