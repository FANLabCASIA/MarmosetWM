```python
species = ['chimp', 'macaque', 'marmoset']

fig = plt.figure(figsize=(10, 2), dpi=200)
fig.tight_layout() #
plt.subplots_adjust(wspace=0.4, hspace=0.4)

fiber1 = 'ALL'
fiber2 = 'AF'

for i,f in enumerate(species):
    
    # print(i,f)
    
    d1 = np.loadtxt(f'{WD}/minKL_loo/minKL_human_{f}_L_{fiber1}.txt')
    d2 = np.loadtxt(f'{WD}/minKL_loo/minKL_human_{f}_L_{fiber2}.txt')
    d = np.concatenate((d1[:,np.newaxis],d2[:,np.newaxis]),axis=1)
    # print(d1.shape, d2.shape,d.shape)
    df = pd.DataFrame(d, columns=['original', 'knockout'])

    if i==0: ax1 = fig.add_subplot(131)
    if i==1: ax1 = fig.add_subplot(132)
    if i==2: ax1 = fig.add_subplot(133)
    sns.kdeplot(df['original'], shade=False, color='#C4C4C4')
    sns.kdeplot(df['knockout'], shade=False, color='r')
    plt.title(f)

plt.show()
```
